```python
df=pd.read_csv("matches.csv",usecols=['team1','team2','toss_winner','toss_decision','venue','winner'])
df=df.dropna(subset=['winner'])
df=df.reset_index(drop=True)
```

```python
print("shape:",df.shape)
df.head()
```

```python
from sklearn .preprocessing import LabelEncoder
df_encoded=df.copy()
le=LabelEncoder()
col_to_encode=['team1','team2','toss_winner','toss_decision','venue','winner']
```

```python
df.tail(10)
```

```python
for col in col_to_encode:
    df_encoded[col]=le.fit_transform(df_encoded[col])

print(df_encoded.tail(10))
df_encoded
```

```python
from sklearn.model_selection import train_test_split
X=df_encoded.drop('winner',axis=1)
y=df_encoded['winner']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=42)

print("traning set:",X_train.shape)
print("test set:",X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_pred=model.predict(X_test)

accuracy=accuracy_score(y_test,y_pred)
print(f"model accuracy:{accuracy:.2%}")

cm=confusion_matrix(y_test,y_pred)
print("\nconfusion Matrix:")
print(cm)
```
